```
road2 <- filter(road1, road1$n > 8)
```

```
road2 %>% 
 ggplot(aes(x=地區,y=n,fill=地區))+
 geom_col()+
 geom_text(aes(label=n),vjust=-0.25)+
 labs(title="台中市各地區道路比",x="地區",y="數量")+theme(legend.position="none")
```
