```
install.packages("ggplots2")
library(ggplot2)
```

```
FeSCN<- c('0.000025','0.00005','0.000075','0.0001','0.000125')
A<- c(1.254,1.257,1.695,2.092,2.392)
df <- data.frame(FeSCN = FeSCN, A = A)
ggplot(data = df, mapping = aes(x = FeSCN, y = A, group = 1)) + geom_line( colour = "blue")+ geom_point(size = 2, shape = 22,
 colour = "darkred", fill = "red")
```
