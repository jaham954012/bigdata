```
install.packages("ggplot2")
library(ggplot2)
qplot(Position , TotalPoints , data= NBA1920 , color=Team)
qplot(GamesPlayed , TotalPoints , data= NBA1920 , color=Team , facets=Position~.)
```

上圖是以位置及總得分數做點圖
下圖是以出場次數擊位置做點圖

```
ggplot(NBA1920,aes(x= FieldGoalsAttempted,y=TotalPoints,color=Position))+geom_point()
```

以x= FieldGoalsAttempted,y=TotalPoint做散布圖

```
NBA1 <- NBA1920[order(NBA1920$Team,decreasing = T),]
```

```
ggplot(NBA1920,aes(x= FieldGoalsAttempted,y=ThreesAttempted,color=Position))+geom_point()+geom_smooth()
```
